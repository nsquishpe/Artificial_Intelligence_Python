```python
#Importo librerias
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Cambiar cabecera metadata (tipos de datos)
sismos[' date']=pd.to_datetime(sismos[' date'])
sismos['smajax']=pd.to_numeric(sismos['smajax'], errors = 'coerce')
sismos['sminax']=pd.to_numeric(sismos['sminax'], errors = 'coerce')
sismos['strike']=pd.to_numeric(sismos['strike'], errors = 'coerce')
sismos['mo']=pd.to_numeric(sismos['mo'], errors = 'coerce')
sismos['fac']=pd.to_numeric(sismos['fac'], errors = 'coerce')
sismos['mpp']=pd.to_numeric(sismos['mpp'], errors = 'coerce')
sismos['mpr']=pd.to_numeric(sismos['mpr'], errors = 'coerce')
sismos['mrr']=pd.to_numeric(sismos['mrr'], errors = 'coerce')
sismos['mrt']=pd.to_numeric(sismos['mrt'], errors = 'coerce')
sismos['mtp']=pd.to_numeric(sismos['mtp'], errors = 'coerce')
sismos['mtt']=pd.to_numeric(sismos['mtt'], errors = 'coerce')
sismos['str1']=pd.to_numeric(sismos['str1'], errors = 'coerce')
sismos['dip1']=pd.to_numeric(sismos['dip1'], errors = 'coerce')
sismos['rake1']=pd.to_numeric(sismos['rake1'], errors = 'coerce')
sismos['str2']=pd.to_numeric(sismos['str2'], errors = 'coerce')
sismos['dip2']=pd.to_numeric(sismos['dip2'], errors = 'coerce')
sismos['rake2']=pd.to_numeric(sismos['rake2'], errors = 'coerce')
sismos['eventid']=pd.to_numeric(sismos['eventid'], errors = 'coerce')
```

Realice un gráfico de dispersión (scatter) en un misma figura (como el ejemplo que se le facilitó) con 3 gráficos con las siguientes variables:

```python
#Diagramas de dispersión. str1, str2
x = (sismos['str1'])
y =  (sismos['str2'])
plt.rcParams.update({'figure.figsize':(7,5), 'figure.dpi':100})
plt.scatter(x, y)
plt.colorbar()
plt.title('')
plt.xlabel('Str1')
plt.ylabel('Str2')
plt.title('Str1 vs Str2')
plt.show()
```

```python
#Diagramas de dispersión. dip1, dip2
x = (sismos['rake1'])
y =  (sismos['rake2'])
plt.rcParams.update({'figure.figsize':(7,5), 'figure.dpi':100})
plt.scatter(x, y)
plt.colorbar()
plt.title('')
plt.xlabel('Rake1')
plt.ylabel('Rake2')
plt.title('Rake1 vs Rake2')
plt.show()
```
